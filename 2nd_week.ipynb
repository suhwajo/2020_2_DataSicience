실습과제 1_1

```
def insertion_sort(intlist):
  if len(intlist) < 2:
    return intlist
  for i in range(1,len(intlist)):
    for j in range(i,0,-1):
      if inlist[j-1]>inlist[j]:
        inlist[j-1],inlist[j]=inlist[j], inlist[j-1]
  return inlist

inlist = [ 10, 6, 8, 1, 1, 2, 1 ]
print(insertion_sort(inlist))
```

실습과제 1_2

2. 정수 리스트를 정렬하는 합병정렬(merge sort) 알고리즘을 Python으로 구현하라

```
def merge_sort(intlist2):
    if len(intlist2) < 2:
        return intlist2
    mid = len(intlist2) // 2
    left = merge_sort(intlist2[:mid])
    right = merge_sort(intlist2[mid:])

    left_num, right_num, merge_list = 0, 0, []

    while left_num < len(left) and right_num < len(right):
        if left[left_num] < right[right_num]:
            merge_list.append(left[left_num])
            left_num += 1
        else:
            merge_list.append(right[right_num])
            right_num += 1

    merge_list += left[left_num:]
    merge_list += right[right_num:]
    return merge_list
  
intlist2 = [10, 6, 8, 1, 1, 2, 1]
print(merge_sort(intlist2))
```

```
def insertion_sort(intlist):
  if len(intlist) < 2:
    return intlist
  for i in range(1,len(intlist)):
    for j in range(i,0,-1):
      if inlist[j-1]>inlist[j]:
        inlist[j-1],inlist[j]=inlist[j], inlist[j-1]
  return inlist


def merge_sort(intlist2):
    if len(intlist2) < 2:
        return intlist2
    mid = len(intlist2) // 2
    left = merge_sort(intlist2[:mid])
    right = merge_sort(intlist2[mid:])
    left_num, right_num, merge_list = 0, 0, []

    while left_num < len(left) and right_num < len(right):
        if left[left_num] < right[right_num]:
            merge_list.append(left[left_num])
            left_num += 1
        else:
            merge_list.append(right[right_num])
            right_num += 1

    merge_list += left[left_num:]
    merge_list += right[right_num:]
    return merge_list

inlist = [ 10, 6, 8, 1, 1, 2, 1 ]
print('- Insertion Sort : ', insertion_sort(inlist))

intlist2 = [10, 6, 8, 1, 1, 2, 1]
print('- Merge Sort : ', merge_sort(intlist2))
```
